```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
#sns.set_context('talk')
from datetime import date
```

```python
PCA_per_TF = pd.read_csv('../exp/'+str(date.today())+'/PCA_results.tsv', sep='\t')
p_vals = pd.read_csv('../exp/'+str(date.today())+'/p_vals.tsv', sep='\t', index_col='TF')
low_p = p_vals.nsmallest(5, columns='C(organ):C(dev_stage)')
low_p2 = p_vals.loc[p_vals['C(organ):C(dev_stage)']<(0.05/len(p_vals))]


for n in range(len(low_p2)):
    p = sns.relplot(x='dev_stage', y=low_p2.index[n], hue='organ', style='organ', kind='line', data=PCA_per_TF)
    plt.xlabel('Developmental stage (days)')
    plt.ylabel('Log2 expression')
    plt.title(low_p2.index[n])
```
