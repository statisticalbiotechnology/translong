```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
#sns.set_context('talk')
from datetime import date
```

```python
PCA_per_TF = pd.read_csv('../exp/'+str(date.today())+'/single_TF_test.csv')
plot = sns.relplot(x='dev_stage', y='Hnf4a', hue='organ', style='organ', kind ='line', data=PCA_per_TF)
#plot.set(ylim=(-4,7))
plt.xlabel('Developmental Stage')
plt.ylabel('Expression')
plt.title(PCA_per_TF.columns[2])
plt.show()
```
